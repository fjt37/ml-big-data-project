```julia
Pkg.add("DataFrames")
using DataFrames
using Plots
```

```julia
cd("../project_data")
data = readtable("bitcoin_dataset.csv");
head(data)
```

```julia
price = data[:, :btc_market_price].data;
volume = data[:, :btc_trade_volume].data;
date = data[:, :Date].data
normalize!(price); normalize!(volume);
```

```julia
X = [price[3:end-1] price[2:end-2] price[1:end-3] volume[3:end-1] volume[2:end-2] volume[1:end-3]];
y = price[4:end];
w = X \ y
```

```julia
period = 10
rows = 1:period:length(price)
plot(date[4:end], price[4:end])
plot!(date[4:end], X*w)
```
